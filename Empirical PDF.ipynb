```python
# -*- coding: utf-8 -*-
"""
Created on Fri Dec 05 15:29:53 2014

@author: wjh

This code takes a 1-d numpy array as input 
and produces a plot of its empirical probability distribution
estimated with the statsmodels univariate kernel density estimator
"""
import pandas as pd
import IPython
import matplotlib
import matplotlib.pyplot as plt 
matplotlib.rcParams['pdf.fonttype'] = 42
%matplotlib inline
import numpy as np
import statsmodels.api as sm 
```

```python

#READ IN DATA FROM CSV
df = pd.read_csv('..\..\GIS\LIDAR\csv\\drySlope.csv', delimiter=',', header=None)
```

```python
df.head()
```

```python
def kde_fit(array):
    dens = sm.nonparametric.KDEUnivariate(array)
    dens.fit()
    return dens

```

```python
Slope = kde_fit(df[0].values)
```

```python

fig = plt.figure()
plt.plot(Slope.support, Slope.density, lw=1, color='r');

#plt.xscale('log')
plt.xlim([0,110])
plt.xlabel("Hillslope gradient (%) ")
plt.ylabel("Empirical PDF")
plt.title("Dry Creek (Central Belt) \n median = " + str(np.median(df[0].values)) + "\n mean = " + str(np.mean(df[0].values)) + "\n std dev = " + str(np.std(df[0].values)))
#fig.savefig('Figs\dryCreek.pdf', transparent=True, bbox_inches='tight', pad_inches=0.2)
```
